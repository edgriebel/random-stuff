# Test of viewing Jupyter Notebooks in GitHub

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
x = np.linspace(0.0,10.0,5000)
noise = np.random.randn(len(x))
```

```python
y = (0.5 * x) + 5 + noise
```

```python
x_df = pd.DataFrame(x, columns=['X Data'])
y_df = pd.DataFrame(y, columns=['Y'])
data = pd.concat([x_df,y_df], axis=1)
```

```python
data.sample(frac=0.1).plot(kind='scatter', x='X Data', y='Y')
```
